## 리뷰 번역 -> 평가 파이프라인

```python
from transformers import pipeline
from langchain_core.prompts import ChatPromptTemplate, PromptTemplate
from langchain_core.prompts.chat import SystemMessagePromptTemplate, HumanMessagePromptTemplate
from langchain_community.llms import Ollama
from langchain_google_genai import ChatGoogleGenerativeAI
import os
from dotenv import load_dotenv
```

```python

# models tried : 

# mradermacher/English-to-korean-translator-GGUF
# GGUF - 양자 텐서 어쩌구 방식이라 오류 사용할 수 없었음. 

# Helsinki-NLP/opus-mt-tc-big-en-ko
# 쌩뚱맞은 결과가 나옴. 
translator1 = pipeline("translation", model="Helsinki-NLP/opus-mt-tc-big-en-ko")

# Helsinki-NLP/opus-mt-tc-bible-big-mul-mul
# 정확도가 많이 떨어짐
# >>KR<< 로 타겟 언어 표시
translator2 = pipeline("translation", model="Helsinki-NLP/opus-mt-tc-bible-big-mul-mul")


# facebook/mbart-large-50-many-to-many-mmt
# 정확도가 꽤 괜찮음 0.6B parameter 
# src_lang, tgt_lang 으로 src -> tgt 언어 표시
translator3 = pipeline("translation", model = "facebook/mbart-large-50-many-to-many-mmt")
# en_XX, ko_KR

# facebook/m2m100_418M
translator4 = pipeline("translation", model = "facebook/m2m100_418M")
```

```python
prompt = ChatPromptTemplate.from_messages(
    [
        # LLM 에게 role / tone 알려주는 template
        SystemMessagePromptTemplate(
            prompt=PromptTemplate(
                template="당신은 번역가입니다. 영어를 한글로 정확하게 번역해주세요.",
                input_variables=[],
            )
        ),
        # 사용자 input 을 받는 부분
        HumanMessagePromptTemplate(
            prompt=PromptTemplate(template="{input}", input_variables=["input"])
        ),
    ]
)

prompt_text = prompt.format(input="Pedro Pascal leads the team back to the retro-future in colorful new version.")
load_dotenv(".env")  # .env 파일에서 환경 변수 로드
api_key = os.getenv("GOOGLE_API_KEY")  # Google API 키를 안전하게 가져오기
llm = ChatGoogleGenerativeAI(model="gemini-2.0-flash", temperature=0.9)
# LLM 생성
local_llm = Ollama(
    model="gpt-oss:20b",
)
```

```python
src = [
    "Pedro Pascal leads the team back to the retro-future in colorful new version.",
    "David Corenswet and Rachel Brosnahan soar, but the script is kryptonite.",
    "Scarlett Johansson and Jonathan Bailey help the franchise roar back to life.",
    "This musical adaptation hits the right notes, but lacks visual variety.",
    "Ridley Scott sequel is epic old-fashioned movie-making with a star turn from Paul Mescal.",
    "Ana de Armas kills busloads of people in giddy, ruthless, orgy of violence.",
    "Clint Bentley's soft-spoken Western is one of the most quietly beautiful films of the year.",
    "Dwayne Johnson's Oscar effort is missing teeth, but not from getting punched in the face.",
    "Francis Ford Coppola's Megalopolis is bloated and unforgivably dull.",
    "The absolute authority and force of Daniel Day-Lewis carries this movie in the end, and what a pleasure to see his return to the screen.",
]
```

```python
# Helsinki-NLP/opus-mt-tc-big-en-ko
for i in src:
    print(i)
    print(translator1(i))
```

```python
# Helsinki-NLP/opus-mt-tc-bible-big-mul-mul
for i in src:
    print(i)
    print(translator2(">>KR<< "+ i))
```

```python
# gemini 
for i in src:
    prompt_text = prompt.format(input=i)
    translated = llm.invoke(prompt_text).content
    print(i)
    print(translated)
```

```python
# gpt oss 20b 
for i in src:
    prompt_text = prompt.format(input=i)
    translated = local_llm.invoke(prompt_text)
    print(i)
    print(translated)
```

```python
# facebook/mbart-large-50-many-to-many-mmt
# 메모리에 로딩이 되어야 하기 때문에 2회 돌려서 시간 체크
for i in src:
    print(i)
    print(translator3(i, src_lang="en_XX", tgt_lang="ko_KR"))
```

```python
# facebook/m2m100_418M
# 메모리에 로딩이 되어야 하기 때문에 2회씩 돌려서 시간 체크
for i in src:
    print(i)
    print(translator4(i, src_lang="en", tgt_lang="ko"))
```

```python
sentimental = pipeline(
    "sentiment-analysis", model= "nlptown/bert-base-multilingual-uncased-sentiment"
)
```

```python
for i in src:
    translated = translator3(i, src_lang="en_XX", tgt_lang="ko_KR")
    result = sentimental(translated[0]["translation_text"])
    print(translated, result)
```

```python
for i in src:
    translated = translator4(i, src_lang="en", tgt_lang="ko")
    result = sentimental(translated[0]["translation_text"])
    print(translated, result)
```
